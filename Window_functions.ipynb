## **Window functions** 





```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from scipy.special import beta
import scipy as sp
from scipy import signal
from scipy.fftpack import fft, fftshift
```

##Tipos de Janelas 

```python
rectang = signal.boxcar(41)
freqrect = np.linspace(0, 41, len(rectang))

triang = signal.triang(41)
freqtria = np.linspace(0, 41, len(triang))

hamming = signal.hamming(41)
freqham = np.linspace(0, 41, len(hamming))

hann = signal.hann(41)
freqhan = np.linspace(0, 41, len(hann))

blackman = signal.blackman(41)
freqblac = np.linspace(0, 41, len(blackman))

harrisnutt = signal.nuttall(41)
freqharris = np.linspace(0, 41, len(harrisnutt))

kaiser = signal.kaiser(41, beta=5)
freqkaiser = np.linspace(0, 41, len(kaiser))

dolphcheby = signal.chebwin(41, at=100)
freqdolphcheby = np.linspace(0, 41, len(dolphcheby))

plt.figure(figsize=(10,8))
plt.rcParams["font.size"] = "16"
plt.plot(freqrect, rectang, freqtria, triang, freqham, hamming, freqhan, hann,
         freqblac, blackman, freqharris, harrisnutt, freqkaiser, kaiser,
         freqdolphcheby, dolphcheby)
plt.title("Janelas")
plt.ylabel("Amplitude")
plt.xlabel("")
plt.grid()
plt.legend(("Retangular", "Triangular", "Hamming", "Hann", "Blackman", 
            "Harris-Nuttal", "Kaiser Î²=5", "Dolph-Chebyshev"))
plt.savefig("output.pdf") 
plt.savefig("output1.pdf", dpi=400, transparent=False)
plt.show()
```

## EXERCISE 2.1 F) 

Para janela de Kaiser,  $\beta$ altera a forma da janela, podendo atÃ© aproximar outras janelas. 

Ã medida que $\beta$ aumenta, a atenuaÃ§Ã£o relativa do lÃ³bulo lateral diminui e a largura do lÃ³bulo principal aumenta. A Figura abaixo mostra como a atenuaÃ§Ã£o relativa do lÃ³bulo lateral permanece aproximadamente a mesma para um parÃ¢metro $\beta$ fixo conforme o comprimento Ã© variado.

```python
kaiserw = signal.kaiser(40, beta=3)
A = fft(kaiserw, 2048) / (len(kaiserw)/2.0)
freq = np.linspace(-0.5, 0.5, len(A))
response = 20 * np.log10(np.abs(fftshift(A / abs(A).max())))

kaiserw2 = signal.kaiser(40, beta=5)
A2 = fft(kaiserw2, 2048) / (len(kaiserw2)/2.0)
freq22 = np.linspace(-0.5, 0.5, len(A2))
response22 = 20 * np.log10(np.abs(fftshift(A2 / abs(A2).max())))

kaiserwN = signal.kaiser(40, beta=8)
AN = fft(kaiserwN, 2048) / (len(kaiserwN)/2.0)
freqN = np.linspace(-0.5, 0.5, len(AN))
responseN = 20 * np.log10(np.abs(fftshift(AN / abs(AN).max())))

plt.figure(figsize=(10,8))
plt.rcParams["font.size"] = "15"
plt.axis([-0.5, 0.5, -80, 0])
plt.title("Janela de Kaiser")
plt.ylabel("[dB]")
plt.plot(freq, response, freq22, response22, freqN, responseN)
# plt.axhline(-3, -80, 2, color="r", **{"ls":"-."})
plt.grid()
plt.legend(("Kaiser $Î²$=3, L=40", "Kaiser $Î²$=5, L=40", "Kaiser $Î²$=8, L=40"))
plt.show()
```

```python
def equivalent_noise_bandwidth(window):
    #Returns the Equivalent Noise BandWidth (ENBW)
    return len(window) * np.sum(window**2) / np.sum(window)**2

def get_enbw_windows():
    #Return ENBW for all the following windows as a dataframe
    window_names = ["boxcar","barthann","bartlett","blackman","hamming","hann","nuttall","triang"]
    
    df = pd.DataFrame(columns=["Janela", "ENBW(bins)", "ENBW correÃ§Ã£o (3-dB)"])
    for window_name in window_names:
        method_name = window_name
        func_to_run = getattr(signal, method_name) #map window names to window functions in scipy package
        L = 41 #Number of points in the output window (variando L para observar a independencia de Hamming)
        window = func_to_run(L) #call the functions
        
        enbw = equivalent_noise_bandwidth(window) #compute ENBW
        
        df = df.append({"Janela": window_name.title(),"ENBW(bins)":round(enbw,2),"ENBW correÃ§Ã£o (3-dB)": round(10*np.log10(enbw),2)},ignore_index=True)
                       
    return df

df = get_enbw_windows() #call the function
df.head(15) #display dataframe
```
