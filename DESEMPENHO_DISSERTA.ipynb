```python
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

##MAE REAL/DÓLAR and EURO/DÓLAR

```python
labels = ["ARIMA", "ARFIMA", "RW", "MLP", "LSTM", "GRU"]
mae_real_means = [0.180, 0.146, 0.127, 0.079, 0.056, 0.038]
mae_euro_means = [0.084, 0.063, 0.113, 0.065, 0.043, 0.027]

x = np.arange(len(labels))  # the label locations
width = 0.46  # the width of the bars

fig, ax = plt.subplots(figsize=(10,8))
rects1 = ax.bar(x - width/2, mae_real_means, width, label="Erro Absoluto Médio (MAE-Real/Dólar)")
rects2 = ax.bar(x + width/2, mae_euro_means, width, label="Erro Absoluto Médio (MAE-Euro/Dólar)")

# Add some text for labels, title and custom x-axis tick labels, etc.
plt.rcParams['font.size'] = '25'
ax.set_ylabel("Magnitude", fontsize=20)
#ax.set_title('Scores by group and gender')
ax.set_xticks(x)
ax.set_xticklabels(labels, color="black", fontsize=22)
ax.legend(fontsize=20)


def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom', fontsize=15)


autolabel(rects1)
autolabel(rects2)

fig.tight_layout()

plt.show()
```
